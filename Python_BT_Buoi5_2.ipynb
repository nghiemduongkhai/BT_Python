```python
from IPython.display import clear_output
import random

def tao_danh_sach(n):
  danh_sach = [random.randint(1, 100) for _ in range(n)]
  return danh_sach

def in_danh_sach(danh_sach):
  print("Danh sách:", danh_sach)

def in_dao_nguoc(danh_sach):
  print("Danh sách đảo ngược:", danh_sach[::-1])

def sap_xep_danh_sach(danh_sach):
  danh_sach_sap_xep = sorted(danh_sach)
  print("Danh sách đã sắp xếp:", danh_sach_sap_xep)

def tim_phan_tu_lon_nhat(danh_sach):
  phan_tu_lon_nhat = max(danh_sach)
  vi_tri_cuoi = len(danh_sach) - 1 - danh_sach[::-1].index(phan_tu_lon_nhat)
  print("Phần tử lớn nhất:", phan_tu_lon_nhat)
  print("Vị trí phần tử lớn nhất cuối cùng:", vi_tri_cuoi)

def dem_phan_tu(danh_sach):
  x = int(input("Nhập giá trị X: "))
  so_luong = danh_sach.count(x)
  vi_tri = [i for i, val in enumerate(danh_sach) if val == x]
  print("Số lượng phần tử bằng", x, "là:", so_luong)
  print("Các vị trí xuất hiện:", vi_tri)

def la_so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def in_vi_tri_so_nguyen_to(danh_sach):
    print("Vị trí các số nguyên tố:")
    for i, num in enumerate(danh_sach):
        if la_so_nguyen_to(num):
            print(i, end=" ")
    print()

def tim_so_duy_nhat(danh_sach):
    so_duy_nhat = list(set(danh_sach))
    print("Các số duy nhất:", so_duy_nhat)

def liet_ke_tan_suat(danh_sach):
  tan_suat = {}
  for so in danh_sach:
    tan_suat[so] = tan_suat.get(so, 0) + 1
  print("Liệt kê các giá trị và tần suất xuất hiện:")
  for so, lan in tan_suat.items():
    print(f"{so}: {lan} lần")

def tim_doan_con_giam(danh_sach):
    doan_con = []
    i = 0
    while i < len(danh_sach):
        j = i
        while j + 1 < len(danh_sach) and danh_sach[j+1] < danh_sach[j]:
            j += 1
        if j > i:
            doan_con.append(danh_sach[i:j+1])
        i = j + 1
    print("Các đoạn con giảm liên tiếp:", doan_con)

if __name__ == "__main__":
  n = int(input("Nhập số phần tử N: "))
  danh_sach = tao_danh_sach(n)

  while True:
    print("\nMenu:")
    print("1. In danh sách")
    print("2. In đảo ngược danh sách")
    print("3. Sắp xếp danh sách")
    print("4. Tìm phần tử lớn nhất")
    print("5. Đếm số lượng phần tử bằng X")
    print("6. In vị trí số nguyên tố")
    print("7. Tìm số duy nhất")
    print("8. Liệt kê tần suất")
    print("9. Tìm đoạn con giảm liên tiếp")
    print("10. Thoát")
    lua_chon = input("Nhập lựa chọn của bạn: ")

    if lua_chon == '1':
      clear_output()
      in_danh_sach(danh_sach)
    elif lua_chon == '2':
      clear_output()
      in_dao_nguoc(danh_sach)
    elif lua_chon == '3':
      clear_output()
      sap_xep_danh_sach(danh_sach)
    elif lua_chon == '4':
      clear_output()
      tim_phan_tu_lon_nhat(danh_sach)
    elif lua_chon == '5':
      clear_output()
      dem_phan_tu(danh_sach)
    elif lua_chon == '6':
      clear_output()
      in_vi_tri_so_nguyen_to(danh_sach)
    elif lua_chon == '7':
      clear_output()
      tim_so_duy_nhat(danh_sach)
    elif lua_chon == '8':
      clear_output()
      liet_ke_tan_suat(danh_sach)
    elif lua_chon == '9':
      clear_output()
      tim_doan_con_giam(danh_sach)
    elif lua_chon == '10':
      clear_output()
      print("Kết thúc chương trình.")
      break
    else:
      clear_output()
      print("Lựa chọn không hợp lệ !!!")
```
